### Petition Data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json("https://petition.parliament.uk/petitions/241584.json")

total_signed = df['data']['attributes']['signature_count']
backgroud = df['data']['attributes']['background']
country_data = df['data']['attributes']['signatures_by_country']

results_upper = {}
results_lower = {}


for dict in country_data:

    if dict['signature_count'] < 100:
        results_lower[dict['name']] = dict['signature_count']
    else:
        results_upper[dict['name']] = dict['signature_count']
```

```python
df_upper_results = pd.DataFrame(list(results_lower.items()),
                      columns=['country','signature_count'])
df_lower_results = pd.DataFrame(list(results_upper.items()),
                      columns=['country','signature_count'])


df_upper_results.plot.barh(x='country', y='signature_count')
df_lower_results.plot.barh(x='country', y='signature_count')
```
